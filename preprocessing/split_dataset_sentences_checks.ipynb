```python
document_stats_sections = {}

for split, instances in final_dataset_sections.items():
    max_sections = max(len(instance['sections_tokens']) for instance in instances)
    min_sections = min(len(instance['sections_tokens']) for instance in instances)
    avg_sections = sum(len(instance['sections_tokens']) for instance in instances) / len(instances)

    document_stats_sections[split] = {
        'max_sections': max_sections,
        'min_sections': min_sections,
        'avg_sections': avg_sections,
        'num_instances': len(instances),
        'total_sections': sum(len(instance['sections_tokens']) for instance in instances)
    }

for split, stats in document_stats_sections.items():
    print(f"{split} Split:")
    print(f"   Max Sections: {stats['max_sections']}")
    print(f"   Min Sections: {stats['min_sections']}")
    print(f"   Avg Sections: {stats['avg_sections']}")
    print(f"   Number of Instances: {stats['num_instances']}")
    print(f"   Total Sections: {stats['total_sections']}")
    print()
```

```python
document_stats = {}

for split, instances in final_dataset_sentences.items():
    max_sentences = max(len(instance['sentences_tokens']) for instance in instances)
    min_sentences = min(len(instance['sentences_tokens']) for instance in instances)
    avg_sentences = sum(len(instance['sentences_tokens']) for instance in instances) / len(instances)

    document_stats[split] = {
        'max_sentences': max_sentences,
        'min_sentences': min_sentences,
        'avg_sentences': avg_sentences,
        'num_instances': len(instances),
        'total_sentences': sum(len(instance['sentences_tokens']) for instance in instances)
    }

for split, stats in document_stats.items():
    print(f"{split} Split:")
    print(f"   Max Sentences: {stats['max_sentences']}")
    print(f"   Min Sentences: {stats['min_sentences']}")
    print(f"   Avg Sentences: {stats['avg_sentences']}")
    print(f"   Number of Instances: {stats['num_instances']}")
    print(f"   Total Sentences: {stats['total_sentences']}")
    print()
```
