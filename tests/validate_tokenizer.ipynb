```python
import gc
import pickle
import os
import pandas as pd
from transformers import BertTokenizerFast
gc.collect() 
input_file = os.path.join('..', 'data', 'mBERT_tokenized.pkl')
with open(input_file, 'rb') as f:
    dataset = pickle.load(f)
```

```python
def allginment(dataset_split, num_docs_to_check=1):
    print(f"\n{'='*40}")
    print(f"DATASET INSPECTION (Documents: {len(dataset_split)})")
    print(f"{'='*40}")

    for i in range(min(num_docs_to_check, len(dataset_split))):
        doc = dataset_split[i]
        
        keys = list(doc.keys())
        print(f"\n📄 Document [{i}]")
        print(f"   - Keys found: {keys}")

        num_sentences = len(doc['input_ids'])
        len_ids = len(doc['input_ids'][0])
        len_mask = len(doc['attention_mask'][0])
        len_labels = len(doc['labels'][0])

        print(f"   - Sentences in doc: {num_sentences}")
        print(f"   - IDs length:       {len_ids}")
        print(f"   - Mask length:      {len_mask}")
        print(f"   - Labels length:    {len_labels}")

        if len_ids == len_mask == len_labels:
            print("   ✅ Alignment: PERFECT (All lengths match)")
        else:
            print("   ❌ Alignment: ERROR (Mismatched lengths!)")

    print(f"{'='*40}\n")

allginment(dataset['train'], num_docs_to_check=50)
```

```python
def peek_sentence(dataset_split, doc_idx, sent_idx, tokenizer):
    pd.set_option("display.max_columns", None)
    doc = dataset_split[doc_idx]
    input_ids = doc["input_ids"][sent_idx]
    labels = doc["labels"][sent_idx]
    mask = doc["attention_mask"][sent_idx]
    
    tokens = tokenizer.convert_ids_to_tokens(input_ids)
    
    df = pd.DataFrame(
        [tokens, labels, mask],
        index=["Token", "NER_Tag", "Mask"]
    )
    
    print(f"\n🔍 Inspection: Document {doc_idx}, Sentence {sent_idx}")
    print("-" * 50)
    print(df.to_string())
    print("-" * 50)
    
    return df
tokenizer = BertTokenizerFast.from_pretrained("bert-base-multilingual-cased")
sentence = peek_sentence(dataset['train'], doc_idx=52, sent_idx=4, tokenizer=tokenizer)
```
