```python
import pickle, os
import pandas as pd
from IPython.display import display
```

```python
base_path = os.getcwd()
data_dir = os.path.join(base_path, "..", "data")

files = {
    'sentences': 'final_dataset_sentences.pickle',
    'sections':  'final_dataset_sections.pickle',
    'full':      'final_dataset.pickle'
}

datasets = {}
for key, filename in files.items():
    path = os.path.join(data_dir, filename)
    with open(path, 'rb') as f:
        datasets[key] = pickle.load(f)
        
final_dataset_sentences = datasets['sentences']
final_dataset_sections  = datasets['sections']
final_dataset           = datasets['full']
```

```python
pd.set_option('display.max_columns', None)
def visualize_patient_ner(dataset, subset, idx):
    patient_id = dataset[subset][idx]['patient_id']
    print(f"Patient ID: {patient_id} (Index: {idx})")
    for i in range(len(dataset[subset][idx]['sentences_tokens'])):
        sentence = pd.DataFrame([dataset['train'][idx]['sentences_tokens'][i], dataset[subset][idx]['sentences_ner_tags'][i]],index=["tokens", "ner_tags"])
        display(sentence)

visualize_patient_ner(final_dataset_sentences, 'train', 4) #4th record of trainset
```

```python
document_stats = {}

for split, instances in final_dataset_sentences.items():
    max_sentences = max(len(instance['sentences_tokens']) for instance in instances)
    min_sentences = min(len(instance['sentences_tokens']) for instance in instances)
    avg_sentences = sum(len(instance['sentences_tokens']) for instance in instances) / len(instances)

    document_stats[split] = {
        'max_sentences': max_sentences,
        'min_sentences': min_sentences,
        'avg_sentences': avg_sentences,
        'num_instances': len(instances),
        'total_sentences': sum(len(instance['sentences_tokens']) for instance in instances)
    }

for split, stats in document_stats.items():
    print(f"{split} Split:")
    print(f"   Max Sentences: {stats['max_sentences']}")
    print(f"   Min Sentences: {stats['min_sentences']}")
    print(f"   Avg Sentences: {stats['avg_sentences']}")
    print(f"   Number of Instances: {stats['num_instances']}")
    print(f"   Total Sentences: {stats['total_sentences']}")
    print()
```
